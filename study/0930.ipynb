```python
#막대그래프..  bar graph . 범주가 있는 데이터 값을 직사각형의 막대로 표현

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(3)
years = ['2018','2019','2020']
values= [100,400,900]

plt.plot(x, values)
plt.xticks(x,years)

plt.show()

# 연도별로 변환 값을 갖는 데이터를 막대그래프로 표현
# np.arange()범위와 간격에 따라 균일한 값을 갖는 배열을 생성
# bar()함수에 배열 [0,1,2] y축 값에 해당하는 리스트 [100,400,900]을 넣었다.
#xticks()에 x와 years를 입력해주면 x축의 눈금 레이블에 2018 2019 2020이 차례대로 표시된다.
```

```python
x = np.arange(3)
years = ['2018','2019','2020']
values= [100,400,900]



plt.bar(x, values , width=0.6, align='edge',color='green',tick_label=years)

#width : 막대의 너비, 디폴트는 0.8
#align : 틱과 막대의 위치를 조정, 기본값은 center, edge설정하면 왼쪽끝이 tick에 맞춰짐

plt.show()
```

```python
# 수평 막대그래프 그리기

y = np.arange(3)
years = ['2018','2019','2020']
values= [100,400,900]

# plt.barh(y,values)
# plt.yticks(y,years)

plt.barh(y,values,height=-0.6,align='edge',tick_label= years,edgecolor='red',linewidth=3)

#height 막대의 높이 기본값 0.8

plt.show()
```

```python
# 산점도 그리기: 두 변수의 상관 관계를 직교 좌표계의 평면에 점으로 표현하는 그래프

np.random.seed(0)

n= 50
x= np.random.rand(n)    # 1 9 8
y= np.random.rand(n)    # 
a= (30*np.random.rand(n))**2
colors=np.random.rand(n)



plt.scatter(x,y,s=a,c=colors)
# s 는 마커의 크기 지정하고 c마커의 색상을 지정한다.
plt.show()
```

```python
# 컬러맵

np.random.seed(0)

arr=np.random.standard_normal((8,100))


plt.subplot(2,2,1)
plt.scatter(arr[0],arr[1],c=arr[1])
plt.spring()
plt.title('spring')
plt.colorbar()

plt.subplot(2,2,2)
plt.scatter(arr[2],arr[3],c=arr[3])
plt.summer()
plt.title('summer')
plt.colorbar()

plt.subplot(2,2,3)
plt.scatter(arr[2],arr[3],c=arr[3])
plt.autumn()
plt.title('autumn')
plt.colorbar()


plt.subplot(2,2,4)
plt.scatter(arr[2],arr[3],c=arr[3])
plt.winter()
plt.title('winter')
plt.colorbar()


plt.show()
```

```python
a= 2.0* np.random.randn(10000) +1.0  #표준편차 2.0, 평균을 1.0을 갖는 정규분포
b= np.random.standard_normal(10000) #표준 정규분포 난수 발생
c= 20.0*np.random.rand(5000)-10.0

plt.hist(a,bins=100,histtype='step',alpha=0.5)
plt.text(1.7,300,'2.0* np.random.randn(10000) + 1.0', fontdict={'color':'blue','weight':'normal','size':8})



plt.hist(b,bins=50,histtype='stepfilled',alpha=0.5)
plt.text(0.7,550,'np.random.standard_normal(10000)', fontdict={'color':'orange','weight':'normal','size':8})



plt.hist(c,bins=100,histtype='step',alpha=0.5)
plt.text(8,70,'20.0*np.random.rand(5000)-10.0', fontdict={'color':'green','weight':'normal','size':8},
            bbox={'boxstyle':'round','ec':(0.3,1.0,0.5),'fc':(0.8,1.0,0.5),'linestyle':'-.','linewidth':2})
#ec = edgecolor, fc=facecolor


plt.show()
```

```python
# 박스 플롯 - 수치 데이터를 표현하는 방식이다.
#           전체 데이터로부터 얻어진 다섯가지 요약 수치를
#           사용해서 그린다.
#           1. 최소값
#           2. 1사분위 25%
#           3. 2사분위 50%
#           4. 3사분위 75%
#           5. 최대값
```

```python
np.random.seed(0)

data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)



fig,ax=plt.subplots()

ax.boxplot([data_a,data_b,data_c],notch=True)



ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')


plt.show()

```

```python
# Notch - 중앙값(median)의 95% 신뢰 구간을 표현할 수 있다.
```

```python
# 수평 박스 플롯
np.random.seed(0)

data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)



fig,ax=plt.subplots()

ax.boxplot([data_a,data_b,data_c],notch=True,vert=False)



ax.set_xlim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_xlabel('Value')


plt.show()
```

```python
#바이올린 플롯 - 데이터 분포와 범위를 한눈에 보기 쉽게 나타내는 그래프 형식

np.random.seed(0)

data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)



fig,ax=plt.subplots()

ax.violinplot([data_a,data_b,data_c])



ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')


plt.show()
```

```python
np.random.seed(0)

data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)



fig,ax=plt.subplots()

vio=ax.violinplot([data_a,data_b,data_c],showmeans=True)



ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

vio['bodies'][0].set_facecolor('blue')
vio['bodies'][1].set_facecolor('red')
vio['bodies'][2].set_facecolor('green')


plt.show()
```
