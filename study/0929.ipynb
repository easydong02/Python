```python
#파이 그래프 - 원 안에 데이터 각 항목이 차지하는 비율의 크기로 이루어진 그래프이다.
# 각 부채꼴 부분이 파이 조각처럼 생겨서 파이그래프로 부른다.

# 다른 그래프와는 다르게 파이 그래프는 원으로 표현하기 때문에 
# 가로 세로 비율이 1:1로 같아야만 그래프가 제대로 보인다.

fruit= ['사과','바나나','딸기','오렌지','포도']
result=[7,6,3,2,2]
ex_value=(0.1,0,0.1,0,0)

plt.figure(figsize=(5,5))
plt.pie(result,labels=fruit,autopct='%.2f%%',startangle=90,counterclock=False,
        explode=ex_value,shadow=True)
plt.show()
```

```python
#눈금 설정 - ticks
a= np.arange(0,2,0.2)
plt.plot(a,a,'r--',a,a**2,'bo',a,a**3,'-.g')

plt.grid(axis='y',color='red',alpha=0.5,linestyle='--')
plt.xticks([0,1,2],labels=['1gb','2gb','3gb'])
plt.tick_params(axis='x',length=10,direction='in',labelsize=10,labelcolor='red',pad=30)

plt.show()
```
