```python
#상관 관계 분석

# 키와 몸무게, 키와 발의 크기, 교육 수준과 대학 진학률
# 변수들 간의 관계의 성향을 분석

```

```python
import matplotlib.pyplot as plt

height =[100,120,130,140,150,160,170,180,190]
foot = [200,205,210,220,230,250,270,280,285]

plt.scatter(height,foot)
plt.xlabel('height(cm)')
plt.ylabel('foot(mm)')
plt.show()

```

```python

height = [100,200,300,400,500,600,700,800,900]
temp = [18,17.5,17,16.5,15,14.5,13,12,11]

plt.scatter(height,temp)
plt.xlabel('height(m)')
plt.ylabel('temp(c)')
plt.show()
```

```python
#상관 계수(기호 r) : 상관계수는 -1 <= r <= 1의 값을 가진다.
#수치가 0에 가까울 수록 부 뎐수간에 상관관계가 없음을 의미
# -1의 경우는 음의 상관관계 1의 경우는 양의 상관관계
```
