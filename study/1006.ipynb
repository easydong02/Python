```python
#서울의 가장 더웠던 날은 언제일까? 기온은 얼마나?

f = open('weather.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)
m=0
date = 0
for row in data:
    if row[-1]:
        row[-1]=float(row[-1])
        if row[-1]>m:
            m = row[-1]
            date=row[0]
    else:
        row[-1]=-999
        

print('서울 기온이 가장 높았던 날은 {}로 {}도 였습니다'.format(date,m))



f.close()

#빈칸을 어떤 실수로 바꾸어야 할지 몰라서 오류 발생


```

```python
f = open('weatherByDay.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)
birth_date='02-02'
result=[]
result1=[]
for row in data:
    if row[-1]:
        row[-1]=float(row[-1])
        row[-2]=float(row[-2])
        if row[0][5:]==birth_date:
            result.append(row[-1])
            result1.append(row[-2])


plt.plot(result,'r')
plt.plot(result1,'b')

f.close()
```

```python
f = open('weatherByDay.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)

result=[]
result1 =[]
for row in data:
    if row[-1]:
        row[-1]=float(row[-1])
        if row[0].split('-')[1]=='08':
            result.append(row[-1])
        elif row[0].split('-')[1]=='01':
            result1.append(row[-1])
    else:
        row[-1]=-999

plt.figure()
plt.hist(result,bins=100,label='aug')
plt.hist(result1,bins=100,label='jan')
plt.legend()

f.close()
```

```python
#버그 고치기..
f = open('weatherByDay.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)

result=[[],[],[],[],[],[],[],[],[],[],[],[]]
i=1
while i<13:
    for row in data:
        if row[-1]:
            if int(row[0].split('-')[1])==i:
                result[i-1].append(float(row[-1]))
    i+=1
print(result)
plt.boxplot(result)

f.close()
```

```python
f = open('people.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)


result =[]

for row in data:
    if '신도림' in row[0]:
        for i in row[3:]: #3번 인덱스부터가 나이 데이터
            result.append(int(i))


plt.plot(result)
f.close()
```

```python
f = open('gender.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)

m =[]
w =[]

for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3]))
            w.append(int(row[-1-i]))
w.reverse()

plt.figure()
plt.barh(range(101),m)
plt.barh(range(101),w)
plt.show()
```

```python
f = open('gender.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)

area=input("원하는 지역을 입력하세요")
m =[]
w =[]

for row in data:
    if area in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3].replace(',','')))
            w.append(int(row[-1-i].replace(',','')))
        break

w.reverse()

plt.figure()
plt.barh(range(101),m,label='남자')
plt.barh(range(101),w,label='여자')
plt.legend()
plt.show()
```

```python
f = open('gender.csv','r',encoding='utf-8')
data= csv.reader(f,delimiter=',')
header=next(data)

area=input("원하는 지역을 입력하세요")
m=0
w=0
for row in data:
    if area in row[0]:
        
        for i in range(0,101):
            m +=int(row[i+3].replace(',',''))
            w +=int(row[-1-i].replace(',',''))
        break

plt.pie([m,w],labels=['남자','여자'],autopct='%.2f%%')
plt.show()

f.close()
```
